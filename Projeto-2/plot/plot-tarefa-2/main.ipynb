```python
import pandas as pd
import matplotlib.pyplot as plt

caminho = '/home/jvloliveira/root/Graduação/Semestre 2 - 2025/IntroFiscomp/Projeto-2/'

# Lista de arquivos e rótulos
arquivos = [
    (caminho+"tarefa-2/saida-1-12694394.txt", "p=0.50"),
    (caminho+"tarefa-2/saida-2-12694394.txt", "p=0.33"),
    (caminho+"tarefa-2/saida-3-12694394.txt", "p=0.25"),
    (caminho+"tarefa-2/saida-4-12694394.txt", "p=0.20"),
]

# Paleta de cores (azul, verde, amarelo, vermelho)
cores = ["blue", "green", "yellow", "red"]

fig, axs = plt.subplots(2, 2, figsize=(12, 8),dpi=600)

for ax, (caminho, label), cor in zip(axs.flat, arquivos, cores):
    df = pd.read_csv(caminho, header=None, sep=',')
    
    # Normalização
    max_value = df[1].max()
    min_value = df[1].min()
    df[1] = (df[1] - min_value) / (max_value - min_value)

    # Plot
    ax.bar(df[0], df[1], alpha=0.7, color=cor, label=label)
    ax.set_xlabel("Posição")
    ax.set_ylabel("Número de caminhantes (normalizado)")
    ax.legend()

plt.savefig("distribuicao_caminhantes_subplots.png")
plt.tight_layout()
plt.show()
```
