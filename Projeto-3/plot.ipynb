```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.exp(2*x*x) * np.tanh(2*x)

def f_prime(x):
    return 4*x*np.exp(2*x**2)*np.tanh(2*x) + 2*np.exp(2*x**2)*(1 - np.tanh(2*x)**2)

def f_double_prime(x):
    term1 = (1/np.cosh(2*x))**2 * (4*x - 2*np.tanh(2*x))
    term2 = (1 + 4*x**2) * np.tanh(2*x)
    return 4 * np.exp(2*x**2) * (term1 + term2)
xdata = np.linspace(-2, 2, 1000)

plt.figure(figsize=(14, 8), dpi=600)
plt.plot(xdata, f(xdata), color='blue', label=r'$f(x) = e^{2x^2} \cdot \tanh(2x)$', linewidth=2.5)
plt.plot(xdata, f_prime(xdata), color='red', label=r"$f'(x)$", linewidth=2)
plt.plot(xdata, f_double_prime(xdata), color='green', label=r"$f''(x)$", linewidth=2)

plt.grid(True, alpha=0.3)
plt.title("Função f(x) e suas derivadas primeiras e segundas", fontsize=14)
plt.xlabel("x", fontsize=12)
plt.ylabel("f(x)", fontsize=12)
plt.ylim(-10, 10)
plt.legend(fontsize=11)
plt.xlim(-1.5, 1.5)

plt.axhline(0, color='black', linewidth=0.8, ls='-')
plt.axvline(0, color='black', linewidth=0.8, ls='-')

plt.savefig("tarefa-1-graf-1.png", dpi=600, bbox_inches='tight')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.exp(2*x*x) * np.tanh(2*x)

def f_prime(x):
    return 4*x*np.exp(2*x**2)*np.tanh(2*x) + 2*np.exp(2*x**2)*(1 - np.tanh(2*x)**2)


def f_double_prime(x):
    term1 = (1/np.cosh(2*x))**2 * (4*x - 2*np.tanh(2*x))
    term2 = (1 + 4*x**2) * np.tanh(2*x)
    return 4 * np.exp(2*x**2) * (term1 + term2)

xdata = np.linspace(-2, 2, 1000)

# Calcular os valores em x = 0.5
x_point = 0.5
y_f = f(x_point)
y_f_prime = f_prime(x_point)
y_f_double_prime = f_double_prime(x_point)

plt.figure(figsize=(14, 8), dpi=600)
plt.plot(xdata, f(xdata), color='blue', label=r'$f(x)$', linewidth=2.5)
plt.plot(xdata, f_prime(xdata), color='red', label=r"$f'(x)$", linewidth=2)
plt.plot(xdata, f_double_prime(xdata), color='green', label=r"$f''(x)$", linewidth=2)

# Marcar os pontos em x = 0.5
# plt.scatter(x_point, y_f, color='blue', s=100, zorder=5, 
        #    label=rf'$f(0.5) = {y_f:.3f}$')

plt.scatter(x_point, y_f_prime, color='red', s=100, zorder=5, 
           label=rf"$f'(0.5) = {y_f_prime:.3f}$")
plt.scatter(x_point, y_f_double_prime, color='green', s=100, zorder=5, 
           label=rf"$f''(0.5) = {y_f_double_prime:.3f}$")

# Adicionar linha vertical em x = 0.5
plt.axvline(x=x_point, color='gray', linestyle='--', alpha=0.7, linewidth=1)

plt.grid(True, alpha=0.3)
plt.title("Função f(x) e suas derivadas primeiras e segundas", fontsize=14)
plt.xlabel("x", fontsize=12)
plt.ylabel("f(x)", fontsize=12)
plt.ylim(-10, 12)
plt.legend(fontsize=11)
plt.xlim(-1.5, 1.5)

plt.axhline(0, color='black', linewidth=0.8, ls='-')
plt.axvline(0, color='black', linewidth=0.8, ls='-')

plt.savefig("tarefa-1-graf-2.png", dpi=600, bbox_inches='tight')
plt.show()

# Imprimir os valores exatos
print(f"f(0.5) = {y_f}")
print(f"f'(0.5) = {y_f_prime}")
print(f"f''(0.5) = {y_f_double_prime}")
```
